```python
 # Now lets spilt the data into train and test so that we can train our model

from sklearn.model_selection import train_test_split
```

```python
X = df.drop(60, axis=1)
y = df[60]
```

```python
X.head(4)
```

```python
y
```

```python
# As we can see that the target variable column have 2 vaariables so we need to hot encode it

y = pd.get_dummies(y)
# this will create R---->1 and M---->0

```

```python
y.head(4)
```

```python
# But since it is a binary classifiaction problem we can drop one of them
# Lets drop column M

y = y.drop(["M"], axis=1)
```

```python
y.head(4)
```

```python
# This can be achieved by one line as well by using y=pd.get_dummies(y, drop_first=True)
```

```python
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print("size of X_train",len(X_train))
print("size of X_test",len(X_test))
print("size of y_train",len(y_train))
print("size of y_test",len(y_test))
```

```python
print("shape of X_train", X_train.shape)
print("shape of X_test",X_test.shape)
```

```python
# This means we have 60 neurons/input features and we will keep decreasing it in the middle layers
```

```python


model = keras.Sequential([
    keras.layers.Dense(60, input_shape=(60,), activation="relu"),
    keras.layers.Dense(30,  activation="relu"),
    keras.layers.Dense(15,  activation="relu"),
    keras.layers.Dense(10,  activation="relu"),
    keras.layers.Dense(5,  activation="relu"),
    keras.layers.Dense(1,  activation="sigmoid")
])
```

```python
model
```

```python
model.compile(optimizer="adam",
             loss = "binary_crossentropy",
             metrics = ["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
ans_sample = model.predict(X_test)
```

```python
ans_sample[:10]
```

```python
def clean_prediction(x):
    ans = []
    for i in x:
        if(i<0.5):
            ans.append(False)
        else:
            ans.append(True)
    return ans
        
```

```python
get_ans = clean_prediction(ans_sample)
```

```python
get_ans[:10]
```

```python
y_test[:10]
```

```python
# Lets do a classification report 

from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, get_ans))
```

```python
# Now lets introduce Dropout Regularization
# So we will basically train the model again but this time we will keep droping 50% of the layers
# The best practice is to drop 50% layer after each layer
```

```python
model2 = keras.Sequential([
    keras.layers.Dense(60, input_shape=(60,), activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,  activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,  activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(10,  activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(5,  activation="relu"),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,  activation="sigmoid")
])


model2.compile(optimizer="adam",
             loss = "binary_crossentropy",
             metrics = ["accuracy"])

model2.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model2.evaluate(X_test, y_test)
```
